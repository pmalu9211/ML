```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.arange(10)
Y = (X - 5)**2 + 3
```

```python
plt.style.use("seaborn")
plt.plot(X,Y)
# plt.plot([11,55])
plt.show()
```

We will implement gradient decent here

```python
x = 10
# setting any random value of x and learning rate for the gradient descent method
lr = 0.2
for i in range(50):
  y = (x - 5)**2 + 3
  plt.scatter(x,y)
  x = x - lr*(2*(x-5))
  print(x)

plt.style.use("seaborn")
plt.plot(X,Y)
# plt.plot([11,55])
plt.show()
```
