```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread("kids.jpg")



# Step 2: Extract color channels
g = img[:, :, 1]
r = img[:, :, 0]
b = img[:, :, 2]

# Step 3: Create a new image array with desired shape and type
ans = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)

ans[:, :, 0] = b  # Assign red channel to the new image
ans[:, :, 1] = g  # Assign green channel to the new image
ans[:, :, 2] = r

print(ans.shape)

plt.imshow(ans)
```

```python
model = cv2.CascadeClassifier("haarcascade_frontalface_alt.xml")
```

```python
faces = model.detectMultiScale(ans,1.1,1)
faces
```

```python
for f in faces:
    print(f)
    x,y,w,h  = f
    cv2.rectangle(ans,(x,y),(x+w,y+h),(0,255,0),2)

plt.imshow(ans)
plt.show()
```
